## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

# Load the starter anonymized dataset
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

# Filter dataset based on standard criteria (matured pages with search presence)
df_filtered = df[(df["impressions_90d"] > 0) & (df["content_age_days"] >= 90)].copy()

total_pages = len(df_filtered)

# Number 1: Total scale of declining pages
declining_pages = len(df_filtered[df_filtered["trend_direction"] == "down"])
declining_pct = (declining_pages / total_pages) * 100

# Number 2: High-demand pages in decline (high impact opportunity)
high_demand_declining = len(
    df_filtered[(df_filtered["trend_direction"] == "down") & (df_filtered["impressions_90d"] >= 500)]
)

# Number 3: Freshness risk (pages not updated in > 180 days with solid impressions)
stale_visible_pages = len(
    df_filtered[(df_filtered["days_since_last_update"] >= 180) & (df_filtered["impressions_90d"] >= 500)]
)
stale_visible_pct = (stale_visible_pages / total_pages) * 100

# Print the verifiable real numbers
print(f"Total analyzed pages: {total_pages:,}")
print(f"1. Declining pages: {declining_pages:,} ({declining_pct:.1f}%)")
print(f"2. High-demand declining pages (>=500 imp): {high_demand_declining:,}")
print(f"3. Stale high-visibility pages (>=180d untouched, >=500 imp): {stale_visible_pages:,} ({stale_visible_pct:.1f}%)")
```

Key Data Evidence Supporting This Lane:

1. Problem Scale: Out of the analyzed portfolio of 30,000 pages, 16,262 pages (54.2%) are currently in a downward performance trend.
2. High-Impact Opportunity: There are 9,961 declining pages that still retain strong demand (>= 500 impressions in 90 days), representing immediate high-value targets for content recovery.
3. Actionable Driver (Freshness Decay): We identified 17 high-visibility pages (>= 500 impressions) that have gone over 180 days without an update (0.1%), representing an initial subset of neglected, high-performing content.
